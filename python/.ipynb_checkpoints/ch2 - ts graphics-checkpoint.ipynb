## Lag plots

```python
aus_production['Quarter_formatted'] = pd.to_datetime(aus_production.Quarter.str.replace(' ', '-'))
aus_production.head()
```

```python
recent_production = aus_production[aus_production.Quarter_formatted.dt.year >= 1992]
```

```python
recent_production.head()
```

```python
lags = {}
for i in range(1, 10):
    lags['B' + str(i)] = recent_production.Beer.shift(i)
```

```python
lagged_df = pd.DataFrame.from_dict(lags).merge(
    recent_production[['Beer', 'Quarter_formatted']], left_index=True, right_index=True
                      )

lagged_df['Quarter_formatted'] = lagged_df.Quarter_formatted.dt.quarter
```

```python
lagged_df.head()
```

```python
beer_lag_long = lagged_df.melt(id_vars=['Beer', 'Quarter_formatted'], value_name='Production', var_name='lag')
p = sns.FacetGrid(beer_lag_long, col='lag', col_wrap=3, hue='Quarter_formatted')
p.map_dataframe(plt.scatter, x='Production', y='Beer')
plt.legend(bbox_to_anchor=(1.5,2), title='Quarter')
```
